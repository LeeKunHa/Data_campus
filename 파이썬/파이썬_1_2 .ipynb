### 반복문_while

- #### while

```python
a=[1,3,9,5]

i=0 #인덱스
while i<len(a):
    print("index: ",i,"value: ",a[i])
    i=i+1
```

```python
a=[1,3,9,5]

i=0 #인덱스
while i<len(a):
    if a[i]>5:
        print("index: ",i,"value: ",a[i])
    i=i+1
```

- #### 무한루프

```python
while True:
    print("end")
```

- #### break

```python
a=[1,3,9,5]

i=0
while i <len(a):
    if a[i]>5:
        break
    print("index: ",i,"value: ",a[i])
    i=i+1
```

### 반복문_for

- #### for

```python
a=[1,3,9,5]

for i in a:
    print(i)
print("end")
```

```python
a=[1,3,9,5]

for i in a:
    print(i, i*2)
print("end")
```

```python
a=[1,3,9,5]

for index,num in enumerate(a):
    print(index,num)
print("end")
```

```python
a=[1,3,9,5]

for index, num in enumerate(a):
    print("index: ",index, num)
print("end")
```

```python
a={"A":"apple", "B":"banana", "C":"carrot"}

for key, values in a.items():
    print(key,values)
```

- #### break

```python
a=[1,3,9,5]

for num in a:
    if num>5:
        break
    print(num)
    
print("end")
```

- #### 중첩

```python
a=[1,3,9,5]
for i in a:
    for j in a:
        print(i,"*",j)
print("end")
```

- #### 구구단을 2 - 9단까지 출력하세요. 

```python
for i in range(2,10):
    print(i,"단")
    for j in range(1,10):
        print(i,"*",j,"=",i*j)
```

- #### 1 - 100 까지 정수 중 3의 배수 또는 7의 배수를 모두 출력하세요.

```python
for i in range(1,101):
    if i%3==0:
        print(i)
    if i%7==0:
        print(i)
```

- #### a = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28] 에서 최대값, 최소값, 평균을 구하세요. 

```python
a = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]
a.sort()
s=len(a)

print("최대값:",a[s-1])
print("최소값:",a[0])
sum=0
for i in range(s):
    sum+=a[i]
print("평균:",sum/s)
```
